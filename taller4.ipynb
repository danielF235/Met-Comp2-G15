```python
x_grid = np.linspace(-1.1,1.1, 500)
y_grid = np.linspace(-1.1,1.1, 500)
x, y = np.meshgrid(x_grid, y_grid)
condition = x**2 + y**2 <= 1
values = np.zeros(x.shape)
values[condition] = np.random.random(size=x[condition].shape)
values[~condition] = np.sin(7*np.arctan2(y[~condition], x[~condition]))
plt.figure(figsize=(8, 6))
plt.imshow(values, extent=[x_grid.min(), x_grid.max(), y_grid.min(), y_grid.max()], origin='lower')

plt.colorbar(label=r"$\phi$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
