```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def system(t,y,omega_0,drag,craro):
    x,v = y
    return np.array([v, -omega_0**2*x -drag*v -craro*np.power(x,3)])

sol = solve_ivp(system, y0=np.array([1.,0.]), t_span=(0,100), args=[1.5,0.5,0.5], max_step=0.1)
```

```python
plt.plot(sol.t, sol.y[0], ".-")
```
